```python
def dot(img,ori,destSize,i,j,radius):
    coefficient = int(destSize/ori.shape[0])
    row = i*coefficient
    col = j*coefficient
    cv2.circle(img,(row+radius,col+radius), radius, (0,0), -1)
```

```python
def dotText(destSize, gray, radius,whitten=False):
    blank = np.ones((destSize,destSize),np.uint8)
    blank = 255*blank
    for i,row in enumerate(gray):
        for j,col in enumerate(row):
            if(gray[j][i]) <200:
#                 if (whitten and i%2==0):
#                     pass
#                 else:
                    dot(blank,gray,destSize,i,j,radius)
    blank = cv2.resize(blank,(80,80))
    blank = cv2.resize(blank,(224,224))
    plt.figure(figsize=(10,10))
    plt.imshow(blank, 'gray')
    plt.show()
    return blank
```
